```python
#pip install pandas geopandas matplotlib shapely

import pandas as pd
import numpy as np
from shapely.geometry import Point
#import geopandas as gpd

# my curated data, without rating data
all_routes = pd.read_pickle('Curated_OpenBetaAug2020_RytherAnderson.pkl.zip')

# load .zip file directly
all_routesB = pd.read_json("./openbeta-usa-routes-aug-2020.zip", lines=True)
```

```python
#This is something that seems to extract the lat long
all_routesB=all_routesB.join(pd.json_normalize(all_routes['metadata']).add_prefix("metadata_")).drop(['metadata'], axis=1)

all_routesB.head(10)
```

```python
#Let me try the same thing to turn the grade into columns too
all_routes=all_routes.join(pd.json_normalize(all_routes['grade']).add_prefix("grade_")).drop(['grade'], axis=1)
all_routes=all_routes.join(pd.json_normalize(all_routes['type']).add_prefix("type_")).drop(['type'], axis=1)

all_routes.sample(2)
```

```python
#Check out the data

print(all_routes.size)
print(all_routes.ndim)
print(all_routes.dtypes)
print(all_routes.shape)
print(all_routes.count())
```

### 2

```python
#Drop some vars you don't need
all_routes=all_routes.drop(all_routes.filter(regex='grade_F').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='grade_F').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='grade_E').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='grade_U').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='grade_Z').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='grade_B').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='grade_yds_aid').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='safety').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='protection').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='description').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='location').columns, axis=1)
all_routes=all_routes.drop(all_routes.filter(regex='left').columns, axis=1)

print(all_routes.count())
all_routes.sample(2)
```
